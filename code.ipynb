```python
# Obtendo uma visão estatistica da base
# Importante pra analisar se a base tem outliers
data.describe()

# count: Número de valores não nulos (entradas) em cada coluna.
# mean: Média (valor médio) dos dados na coluna.
# std: Desvio padrão (uma medida de dispersão) dos dados na coluna.
# min: Valor mínimo na coluna.
# 25%: O primeiro quartil (valor abaixo do qual 25% dos dados caem).
# 50% (mediana): O segundo quartil (valor abaixo do qual 50% dos dados caem).
# 75%: O terceiro quartil (valor abaixo do qual 75% dos dados caem).
# max: Valor máximo na coluna.
```

```python
# Importando as bibliotecas de graficos
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Visualizando as contagens
sns.countplot(x='sex', data=data) #sex
plt.show()

sns.countplot(x='children', data=data) #filhos
plt.show()

sns.countplot(x='region', data=data) # região
plt.show()

sns.countplot(x='smoker', data=data) # fumantes
plt.show()

# Visualizando a distribuição
sns.histplot(data['age'], kde=False) # qual a distribuição de idades
plt.show()

sns.histplot(data['bmi'], kde=False) # qual a distribuição de imc 
plt.show()

sns.histplot(data['charges'], kde=False) # qual a distribuição de custos
plt.show()

```

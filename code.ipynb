```python
# Importando as bibliotecas de graficos
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Visualizando as contagens
sns.countplot(x='sex', data=data) #sex
plt.show()

sns.countplot(x='children', data=data) #filhos
plt.show()

sns.countplot(x='region', data=data) # regi√£o
plt.show()

sns.countplot(x='smoker', data=data) # fumantes
plt.show()

# Visualizando a distribui√ß√£o
sns.histplot(data['age'], kde=False) # qual a distribui√ß√£o de idades
plt.show()

sns.histplot(data['bmi'], kde=False) # qual a distribui√ß√£o de imc 
plt.show()

sns.histplot(data['charges'], kde=False) # qual a distribui√ß√£o de custos
plt.show()

```
